## Linear filtering by convolution (mean filter)

**<a id="#">Task 3.1.1.</a>** Load the image `data/lena.png` and show it:

```python
img = plt.imread('data/lena.png')
plt.imshow(img)
```

**<a id="#">Task 3.1.2.</a>** Implement a mean filter:

```python
def meanfilter(img, size):
    result = numpy.zeros(img.shape)
    n = (size - 1) // 2
    for p in numpy.ndindex(img.shape):
        neighborhood_sum = 0
        for q in numpy.ndindex(size, size):
            neighborhood_sum = neighborhood_sum + img[p[0]-n+q[0]-n,p[1]-n+q[1]-n]
        result[p] = neighborhood_sum / size
    return result
```

```python
plt.imshow(meanfilter(img, 5))
```

```python
meanimg = plt.imread('data/lena_meanfilter5.png')
plt.imshow(meanimg)
```
